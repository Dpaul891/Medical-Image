```python
import os
from PIL import Image
import numpy as np
from torchvision.transforms import functional as F
from tqdm import tqdm
```

```python
train_ls = []
val_ls = []
with open('./COVID-19_Dataset/ImageSets/Segmentation/train_mask.txt', 'r') as f:
    content = f.read().splitlines()
    for contenti in content:
        train_ls.append(contenti.split('/')[-1])
with open('./COVID-19_Dataset/ImageSets/Segmentation/val_mask.txt', 'r') as f:
    content = f.read().splitlines()
    for contenti in content:
        val_ls.append(contenti.split('/')[-1])
with open('./COVID-19_Dataset/ImageSets/Segmentation/train.txt', 'w') as f:
    for i in train_ls:
        f.write(i + '\n')
with open('./COVID-19_Dataset/ImageSets/Segmentation/val.txt', 'w') as f:
    for i in val_ls:
        f.write(i + '\n')
```

```python
path = './COVID-19_Dataset/JPEGImages'
for name in tqdm(os.listdir(path)):
    img_path = os.path.join(path, name)
    img = Image.open(img_path)
    img = F.resize(img, 256)
    img = img.convert('RGB')
    img.save(img_path)
```

```python
img = Image.open('./COVID-19_Dataset/JPEGImages/COVID-1.png')
print(img.mode)
img = np.array(img)
img.shape
```

```python
path = './COVID-19_Dataset/SegmentationClass'
for name in tqdm(os.listdir(path)):
    img_path = os.path.join(path, name)
    img = Image.open(img_path)
    img = np.array(img)
    img = img[:, :, 0]
    img[img == 255] = 1
    img = Image.fromarray(img, mode='P')
    img.save(img_path, mode='P')
```
